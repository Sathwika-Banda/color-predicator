```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
def generate_data(num_samples=1000):
    colors = np.random.rand(num_samples, 3)
    labels = np.random.rand(num_samples, 3)
    return colors, labels
x_train, y_train = generate_data()
model = keras.Sequential([
    layers.Dense(32, activation='relu', input_shape=(3,)),  # Input layer
    layers.Dense(16, activation='relu'),  # Hidden layer
    layers.Dense(3, activation='sigmoid')  # Output layer
])
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=50, batch_size=32)
x_test, y_test = generate_data(num_samples=10)
predictions = model.predict(x_test)
for i in range(10):
    plt.figure(figsize=(5, 2))
    plt.subplot(1, 2, 1)
    plt.imshow([[x_test[i]]])
    plt.title("Input Color")
    plt.subplot(1, 2, 2)
    plt.imshow([[predictions[i]]])
    plt.title("Predicted Color")
    plt.show()
```
